```python
people = ['VARSHA','SAM','TEENA','HEMA','SWETHA','SINDHU',
 'AKSHITHA','HARINI','SATHWIKA','SAHITHI','PRASHAMSA','VYSHNAVI']
age = [21,12,32,45,37,18,28,52,5,40,48,15]
weight = [55,35,77,68,70,60,72,69,18,65,82,48]
height = [160,135,170,165,173,168,175,159,105,171,155,158]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight, height)
plt.title("Relationship between Weight and Height of FRIENDS")
plt.ylabel("Height in Meters")
plt.xlabel("Weight of FRIENDS in Kgs")
plt.show()
```

```python
plt.figure(figsize=(7,4))
plt.title("People's weight in kgs",fontsize=16,
 fontstyle='italic')
# Main plot function 'bar'
plt.bar(x=people,height=weight, width=0.6,
 color='blue',edgecolor='k',alpha=0.6)
plt.xlabel("People",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("Weight (in kgs)",fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(3,2))
# Main plot function 'hist'
plt.hist(weight,color='yellow',edgecolor='k', alpha=0.75,bins=5)
plt.title("Histogram of friends weight",fontsize=13)
plt.xlabel("Weight in kgs",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
