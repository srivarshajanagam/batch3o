```python
people = ['varsha','kiran,sam','sindhu','sathwika','swetha']
age = [19,20,18,19,19,19]
weight = [60,55,35,45,47,46]
height = [160,150,160,150,140,145]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight,height)
plt.title("relationship between height and weight of my friends")
plt.ylabel("height in meters")
plt.xlabel("weight of my friends in kgs")
plt.show()
```

```python
plt.scatter(age,height)
plt.title("relationship between age and height of my friends")
plt.xlabel("age of my friends")
plt.ylabel("height in meters")
plt.show()
```

```python
plt.figure(figsize=(5,4))
plt.title("my friends weight in kgs",fontsize=16,fontstyle='italic')
#main plot function 'bar'
plt.bar(x=people,height=weight, width=0.6, color='black', edgecolor='k',alpha=0.6)
plt.xlabel("people",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("weight(in kgs)",fontsize=15)
plt.show()
```
